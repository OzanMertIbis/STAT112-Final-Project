```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_excel("Term_Deposit_Cleaned.xlsx") # Import the cleaned data and turn it into a DataFrame
print(df.head()) # Printed the first five rows of the DataFrame to see a concise view of the rows and columns
print()
print(df.info()) # Printed the info (such as row counts, column counts, data types etc.) concerning the DataFrame to better understand it

hue_colours = {"Yes": "#D61A3C",
             "No": "#808080"}

sns.set_style("whitegrid") # Set the default graph style Seaborn's "whitegrid", to make visualizations easier to read
```

## 1) What is the mean age of individuals who opt for a term deposit in comparison to those who do not?


```python
g = sns.catplot(x="Deposit", y="Customers Age", data=df, kind="point", join=False, capsize=0.03, palette=hue_colours) # Constructed a point plot using Deposit and Customers Age columns

g.fig.suptitle("Mean Age of Individuals per Term Deposit Subscription Status", y=1.02) # Added a title

g.set(xlabel="Subscription to Term Deposit", ylabel="Average Customer Age") # Added appropriate labels for the x and y axes

plt.show() # Printed the visualization
```

## 2) What is the disparity in the average account balance between customers who opt for a term deposit and those who do not?


```python
g = sns.catplot(y="Deposit", x="Balance",data = df, kind="box", palette=hue_colours) # Constructed a box plot using Deposit and Balance columns

g.fig.suptitle("Account Balance per Term Deposit Subscription Status", y=1.02) # Added an appropriate title

g.set(ylabel="Subscribed to Term Deposit", xlabel="Account Balance") # Added appropriate labels to the x and y axes

plt.show() # Printed the graph
```

## 3) How are term deposit subscribers distributed among occupations?

```python
job_list = {}
for x in df["Customers Job"]:
  job_list[x] = 0
# Created a dictionary to store the number of occurence for each job

for x in df["Customers Job"]:
  job_list[x] += 1
# Counted the number of occurence for each item in the job_list list

job_list_new = []
for x in df["Customers Job"]:
  if x not in job_list_new:
    job_list_new.append(x)
  else:
    continue
# Created a list of all the unique items under Customers Job column

print(job_list_new) # Printed the new lists and dictionaries we created to get a clear view of what we've done so far before moving on
print(job_list)

deposit_job = df[["Customers Job", "Deposit"]]
deposit_job # Used sub-setting for Customers Job and Deposit columns to get a better view of the data we're working with
```

```python
df_dict = {}
job_list = []
yes_num = []
no_num = []
total = []
for x in job_list_new:
  a = df[df["Customers Job"]==x][df["Deposit"]=="Yes"].count()
  b = df[df["Customers Job"]==x][df["Deposit"]=="No"].count()
  c = df[df["Customers Job"]==x].count()
  job_list.append(x)
  yes_num.append(int(a[0]))
  no_num.append(int(b[0]))
  total.append(int(c[0]))
# Created lists containing the information we need to visualize this graph

df_dict["Customers Job"] = job_list
df_dict["Term Deposit Subscriber"] = yes_num
df_dict["Not a Term Deposit Subscriber"] = no_num
df_dict["Total"] = total
# Combined the lists created into a single dictionary

subs_df = pd.DataFrame(df_dict)  # Transformed the dictionary we created into a DataFrame seperate from the original DataFrame
print(subs_df)
g = sns.catplot(x="Customers Job", y="Term Deposit Subscriber", data=subs_df, kind="bar", errorbar=None, color="#D61A3C")  # Constructed a bar graph to answer the question we asked

g.fig.suptitle("Number of Term Deposit Subscribers per Occupation", y=1.02)  # Added a title and axes labels to the graph
g.set(xlabel="", ylabel="Number of Term Deposit Subscribers")

plt.xticks(rotation=90) # Rotated each item in the x-axis 90 degrees to avoid overlapping
plt.show() # Printed the graph
```

## 4) Is there a noticeable variation in the subscription rates depending on one's marital status? (e.g., married, single, divorced)?



```python
g = sns.catplot(x="Deposit", data=df, kind="count", col="Marital Status", col_order=["Married", "Divorced", "Single"], palette=hue_colours) # Construced a graph to answer the question asked

g.fig.suptitle("Relationship Between Subscription Status and Marital Status", y=1.02) # Added a title

g.set(xlabel="Subscribed to Term Deposit", ylabel="Number of Customers") # Added axes labels

plt.show() # Printed the graph
```

## 5) What is the distribution of age among customers?

```python
g = sns.displot(df["Customers Age"], kde=True, bins=7, color="#D61A3C", alpha=0.7) # Construced a histogram of the customers' ages. Created 7 classes and adjusted transparency

g.fig.suptitle("Distribution of Age Among Customers", y=1.02) # Added a title

g.set(xlabel="Customers' Age", ylabel="Number of Customers") # Added label titles

plt.show() # Printed the graph
```

## 6) How is Contact Duration distributed among customers who have opted in to Term Deposit plans compared to customers who did not?

```python
g = sns.catplot(x="Deposit", y="Duration", data=df, kind="violin", palette=hue_colours) # Construced a violin plot to answer the question

g.fig.suptitle("Contact Duration Rate per Deposit Subscription Status", y=1.02) # Added a title

g.set(xlabel="Subscribed to Term Deposit", ylabel="Contact Duration (in seconds)") # Added label titles

plt.show() # Printed the graph
```

## 7) What is the number of customers with a housing loan who also subscribed to a term deposit, and how does this compare to customers who do not haveahousingloan?

```python
g = sns.catplot(x="Deposit", data=df, kind="count", col="House Loan", palette=hue_colours) # Construced a countplot to answer the quesiton asked

g.fig.suptitle("Relationship Between Term Deposit Subscription Status and Housing Loan Subscription", y=1.02) # Added a title
g.set(xlabel="Subscribed to Term Deposit", ylabel="Number of Customers") # Added axes labels

plt.show() # Printed the graph
```

## 8) Is there a relationship between Income and Loan Amount?

```python
g = sns.scatterplot(x="Income", y="Loan_amount", data=df, hue="Deposit", palette=hue_colours) # Construced a scatterplot to answer the question asked

g.set(xlabel="Income", ylabel="Loan Amount") # Added axes labels

corr_coef = df["Income"].corr(df["Loan_amount"]) # Computed the Correlation Coefficient between these variables to get a better understanding of the linear relationship between them
print(corr_coef)

plt.show() # Printed the graph
```
